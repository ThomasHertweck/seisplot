In this little tutorial, we would like to create a wiggle plot rather than an image plot (see tutorial #1). First, we load all required modules:

```python
import numpy as np

import sys
sys.path.insert(1, '../')

import seisplot
```

Then we load the seismic test data to plot (for details, see tutorial #1).

```python
cmpgather = np.load("data/cmp.npy")
```

The data (a single CMP gather) consists of 101 traces. Compared to tutorial #1 where some of the parameters have already been explained, all we need to do is set the `plottype` parameter to "wiggle":

```python
fig, ax = seisplot.plot(cmpgather, plottype="wiggle", haxis="offset", width=5, height=6,
                        vlabel="Time (s)", hlabel="Offset (m)", 
                        vmajorticks=0.2, vminorticks=0.1, hminorticks=100)
```

Again, that's pretty easy, isn't it?!

You can modify the line color, fill color, transparency, wiggle excursion, and so on, to customize the display. The *plot()* method has quite a lot of parameters:
